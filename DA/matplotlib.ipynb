```python
randn_cumsum=np.random.randn(1000).cumsum() # 1000개의 난수 데이터
```

```python
fig,ax=plt.subplots()
ax.set_title('My first matplotlib plot') # 그래프의 제목
ax.set_xlabel('Stages') # x축 이름
ax.set_ylabel('Random Value') # y축 이름
ax.set_xticks([0, 250, 500,750, 1000]) # x축 눈금
ax.set_xticklabels(['one','two','three','four','five'],rotation=30,fontsize='small') # x축 눈금에 각각 이름 부여, rotation: 출력할 이름 각도, fontsize: 글씨 크기
ax.annotate('250\n250.00\n',  # 출력할 글씨
            xy=(250,randn_cumsum[250]+0.5), # 표시하고자 하는 x,y 위치(화살표가 가리킬 위치:250번째 데이터의 y축 위치에서 0.5위의 자리에)
            xytext=(250,randn_cumsum[250]+10), # 글씨를 출력할 위치(250번째 데이터의 y축 위치에서 +10 자리에)
            arrowprops=dict(facecolor='black',headwidth=4,width=2,headlength=4), # 화살표
            horizontalalignment='right', # 문자열 오른쪽 정렬
            verticalalignment='top') # 문자열 위쪽 정렬
# ax.set_xlim(255, 275) # x 범위
# ax.set_ylim(-10,0) # y 범위
ax.plot(randn_cumsum)
plt.show()
```

```python
randn_cumsum=np.random.randn(500).cumsum() # 1000개의 난수 데이터

fig,ax=plt.subplots()
ax.set_title('My first matplotlib plot') # 그래프의 제목
ax.set_xlabel('Data') # x축 이름
ax.set_ylabel('Value') # y축 이름
ax.set_xticks([0, 250, 500]) # x축 눈금
ax.set_xticklabels(['one','two','three'],rotation=90,fontsize='small') # x축 눈금에 각각 이름 부여, rotation: 출력할 이름 각도, fontsize: 글씨 크기
ax.annotate('Here!',  # 출력할 글씨
            xy=(250,randn_cumsum[250]-0.5), # 표시하고자 하는 x,y 위치(화살표가 가리킬 위치:250번째 데이터의 y축 위치에서 0.5위의 자리에)
            xytext=(250,randn_cumsum[250]-10), # 글씨를 출력할 위치(250번째 데이터의 y축 위치에서 -10 자리에)
            arrowprops=dict(facecolor='red',headwidth=4,width=2,headlength=4), # 화살표
            horizontalalignment='center', # 문자열 오른쪽 정렬
            verticalalignment='center') # 문자열 위쪽 정렬
# ax.set_xlim(255, 275) # x 범위
# ax.set_ylim(-10,0) # y 범위
ax.plot(randn_cumsum)
plt.show()
```

```python
fig=plt.figure()

ax1=fig.add_subplot(2,2,1) # 2행 2열 1번째에 ax1
ax2=fig.add_subplot(2,2,2) # 2행 2열 2번째에 ax2
ax3=fig.add_subplot(2,2,3) # 2행 2열 3번째에 ax3

np_randn=np.random.randn(50)

ax1.hist(np_randn,bins=20,color='k',alpha=0.3) # 데이터, bins: 막대의 개수, color: 막대의 색깔, alpha: 막대 색의 투명도
ax2.scatter(np.arange(50),np.arange(50)+3*np_randn) # 데이터, x축 값, y축 값
ax3.plot(np_randn.cumsum(),'k--') # 데이터, 검은색 점선

plt.show()
```

```python
np_randn=np.random.randn(50)

fig,axes=plt.subplots(2,3) # 2행 3열로 생성

axes[0,0].hist(np_randn,bins=20,color='k',alpha=0.3) # 0행 0열에 히스토그램
axes[0,2].scatter(np.arange(50),np.arange(50)+3*np_randn) # 0행 2열에 산포도
axes[1,1].plot(np_randn.cumsum(),'k--') # 1행 1열에 꺾은선 그래프

plt.show()
```

```python
fig,axes=plt.subplots(2,2)

for i in range(2):
  for j in range(2):
    axes[i,j].hist(np.random.randn(500),bins=50, color='k', alpha=0.5)

fig.subplots_adjust(wspace=2, hspace=2) # 각 axes간 가로 간격, 세로 간격

plt.show()
```

```python
randn1_cumsum=np.random.randn(500).cumsum()
randn2_cumsum=np.random.randn(500).cumsum()
randn13cumsum=np.random.randn(500).cumsum()
```

```python
fig, axes=plt.subplots(2,2)

axes[0,0].set_title('randn1_cumsum')
axes[0,1].set_title('randn2_cumsum')
axes[1,0].set_title('randn3_cumsum')
axes[1,1].set_title('randn4_cumsum')

axes[0,0].plot(randn1_cumsum,'k--') # 검은색 점선
axes[0,1].plot(randn2_cumsum,'k--') # 검은색 점선
axes[1,0].plot(randn13cumsum,'k--') # 검은색 점선

axes[1,1].plot(randn1_cumsum,'r--') # 빨간색 점선
axes[1,1].plot(randn1_cumsum,'b-') # 파란색 점선
axes[1,1].plot(randn1_cumsum,'go-') # 초록색 점선, 마크

axes[0,0].legend()
axes[1,1].legend(loc='best')
# label이 없이 legend를 호출했기 때문에 오류메세지 출력됨. 그래프는 정상적으로 그려짐.

ax.plot(randn_cumsum)

plt.show
```

```python
np_randn=np.random.randn(7)
label_x=list()
for i in range(7):
  label_x.append(str(i))
```

```python
fig,axes=plt.subplots(2,1)

axes[0].set_xticks(range(7))
axes[0].set_xticklabels(label_x)
axes[0].bar(range(7),np_randn,color='k',alpha=0.7) # bar(): 막대그래프

axes[1].set_yticks(range(7))
axes[1].set_yticklabels(label_x)
axes[1].bar(range(7),np_randn,color='k',alpha=0.7)

plt.show()
```

```python
ss=pd.Series(np.random.randn(10).cumsum(),
             index=np.arange(0,100,10))
ax=ss.plot() # 꺾은선 그래프
plt.show()
```

```python
df=pd.DataFrame(np.random.randn(10,4).cumsum(0),  # 10행 4열의 난수의 행방향 누적합계
                columns=['A','B','C','D'],
                index=np.arange(0,100,10))
ax=df.plot()
plt.show()
```

```python
ss=pd.Series(np.random.randn(16),
             index=list('abcdefghijklmnop'))
```

```python
fig,axes=plt.subplots(2,1)

ss.plot.bar(ax=axes[0],color='k',alpha=0.7)
ss.plot.barh(ax=axes[1],color='k',alpha=0.7)

axes[0].legend()
axes[1].legend()

plt.show()
```

```python
df=pd.DataFrame(np.random.rand(6,4),
                index=['one','two','three','four','five','six'],
                columns=pd.Index(['A','B','C','D']))

ax=df.plot.bar()
ax.set_xlabel('Stages')

plt.show()
```

```python
ax=df.plot.barh(stacked=True,alpha=0.5) # 누적 막대 그래프
ax.set_ylabel('Value')

plt.show()
```
