```python
ss=pd.Series([7,-5,7,4,2,0,4])
print(ss.rank())
print(ss.rank(method='first')) # first:동점자가 있을 경우, 먼저나오는 index 부터 순위 부여
print(ss.rank(ascending=False,method='max')) # max:동점자 중 가장 높은 등수 (1등,2등 -> 2등 부여)
print(ss.rank(method='first').sort_values())
```
