## DataFrame 가공

#### 사본 DataFrame 생성

```python
df_new=pd.DataFrame(df_csv,copy=True) # 복사본 만들기
```

#### 새로운 Column 생성

```python
df_new['시도군구명']=df_csv['시도명'].add(' ').add(df_csv['시군구명']) # 시도군구명=시도명+' '+시군구명
```

```python
df_new['시도군구명']
```

#### 겹치지 않는 '시도군구명'열 값 출력

```python
df_new['시도군구명'].unique()
```

```python
len(df_new['시도군구명'].unique()) # 겹치지 않는 시도군구명 307개
```

#### '진구' 문자열을 포함한 '시도군구명'열 값 출력

```python
for i in df_new['시도군구명'].unique():
  if '진구' in i: # 진구가 들어가는 시도군구명 출력
    print(i)
```

#### 값이 일치하는 행 찾기
- 완전 일치하는 행만 찾음

```python
 # 부산광역시진구인 시군구명을 찾기
```

```python
df_new['시군구명']=='부산광역시진구' # 비교연산이기 때문에 T/F결과
```

```python
df_new[df_new['시군구명']=='부산광역시진구'] # True인 인덱스를 찾아 출력
```

#### 문자열을 포함하는 값을 가진 행 찾기
- '시군구명' 열에서 '진구'라는 문자열을 가진 행 추출

```python
# 진구가 들어가는 시군구명 찾기
```

```python
df_new['시군구명'].str.contains('진구') 
```

```python
df_new[df_new['시군구명'].str.contains('진구')]
```

#### 시군구명이 부산광역시진구인 행만 찾아 새로운 DataFrame으로 생성

```python
df_bjd=pd.DataFrame(df_new[df_new['시군구명']=='부산광역시진구'],copy=True)
df_bjd
```

#### 객체 Dataframe 앞 부분 몇 개만 출력

```python
for i in df_bjd.columns[:3]: # column 3개
  for j in df_bjd.index[:3]: # index 3개 
    print(df_bjd[i][j],end=', ')
  print()
```

```python
df_bjd.head() # 앞에서부터 5개 출력
```

#### '부산광역시진구' 문자열을 '부산진구'로 변경
- for 반복문 이용
- 전체 행/열의 값을 탐색하여 변경

```python
for i in df_bjd.columns:
  for j in df_bjd.index:
    if '부산광역시진구' in str(df_bjd[i][j]): # 문자열 '부산광역시진구'가 들어있으면 
      df_bjd[i][j]=df_bjd[i][j].replace('부산광역시진구','부산진구') # 문자열 '부산광역시진구'를 '부산진구'로 바꾸기
```

```python
df_bjd['시도군구명'] # 부산진구로 바뀜
```

```python
df_bjd[df_bjd['시도군구명']=='부산광역시 부산광역시진구'] # 부산광역시진구는 0개 확인! 모두 부산진구로 고쳐짐
```

```python
 df_new['시군구명'][3780] # 원본데이터는 여전히 부산광역시 진구임
```

#### 변경한 DataFrame을 원본 DataFrame에 덮어쓰기

- 메소드 updata는 Columns/Index가 일치하는 위치의 값을 덮어씀

```python
df_final=pd.DataFrame(df_new,copy=True) # df_new 원본데이터 복사본 df_final
df_final.update(df_bjd) # df_bjd를 df_final에 업데이트
```

```python
df_final[df_final['시군구명']=='부산광역시진구'] 
# df_final에 부산광역시진구가 있는지 확인 -> 업데이트했으므로 없음!!!!
```

```python
df_final[df_final['시군구명']=='부산진구'] 
# df_final에 부산진구가 있는지 확인
# 기존 부산진구(69개)와 부산광역시진구->부산진구(43개) 모두 출력됨. 112개
```

```python
print(df_new[df_new['시군구명']=='부산진구'].shape)
print(df_new[df_new['시군구명']=='부산광역시진구'].shape)
print(df_final[df_final['시군구명']=='부산진구'].shape)
```

#### 해당 문자열 사이의 포함 여부

- between: 두 개의 매개변수 사이에 포함하는 문자열을 가진 열을 찾음

```python
df_csv['데이터기준일자'].between('2021-10-01','2021-11-31')
# 범위 안에 있는 문자열을 가진 열을 찾아줌. 문자열도 between 가능하다
```

##### 실습2) '데이터기준이자'가 '2021-10-01'에서 '2021-11-31'사이 인 행만 DataFrame에서 추출
- 조건에 일치하는 '데이터기준일자' 값만 추출

```python
df_csv[df_csv['데이터기준일자'].between('2021-10-01','2021-11-31')]['데이터기준일자']
# 범위에 맞는 데이터의 데이터기준일자만 출력
```

```python
np.sort(df_csv[df_csv['데이터기준일자'].between('2021-10-01','2021-11-31')]['데이터기준일자'].unique())
# 범위에 맞는 데이터의 데이터기준일자 -> 겹치는거는 하나로 모아서 정렬
```

## DataFrame to CSV

#### 객체 DataFrame을 CSV파일로 생성

- 원본 CSV 파일과 Columns를 일치시키기 위해, index 정보를 제외한 내용으로 파일 생성

```python
df_final.to_csv(path+'/새로운.csv',index=False)
```
