```python
from sklearn.preprocessing import StandardScaler

iris_scaler=StandardScaler().fit(df_iris[feature_names]) # fit: 평균, 표준편차 계산
iris_scaler
```

#### 표준 점수 DataFrame 생성 / 그래프 생성

```python
df_iris_scaler=pd.DataFrame(iris_scaler_result,columns=feature_names)
df_iris_scaler['target']=df_iris['target'] # target 열 추가

sns.pairplot(df_iris_scaler,diag_kind='kde',hue='target') # seaborn으로 그리기
plt.show()
```

#### PCA 분석(Principal Component Analysis; 주성분분석)

- 변수(특징, 속성)가 많을수록, 정확한 분석이 가능하다.
- 변수가 많아질수록, 산점도 행렬로 변수간 상관관계 분석이 어려워진다.



- 변수 A가 변화하면 변수 B 또한 변화 : 선형 연관성이 있음


- 변수 B가 변화하면 변수 B는 변화하지 않음 : 선형 연관성이 없음


- 선형연관성이 있을 경우, 이를 가로지르는 제 3 데이터(주성분)를 생성








##### 붓꽃 속성의 PCA 분석

```python
from sklearn.decomposition import PCA

iris_scale_pca=PCA(2) # 2개의 주성분 만들기

iris_pca_trans = iris_scale_pca.fit_transform(df_iris_scaler[feature_names])

print(iris_pca_trans.shape)
```

```python
df_iris['pca_x']=iris_pca_trans[:,0] # 0번째 열의 모든 값 pca_x에 저장
df_iris['pca_y']=iris_pca_trans[:,1] # 1번째 열의 모든 값 pca_y에 저장
```

##### PCA 분석 결과 출력

```python
fig, ax=plt.subplots()
markers=['^','s','o']

for i in range(3):
  x_axis_data=df_iris[df_iris['target']== i]['pca_x']
  y_axis_data=df_iris[df_iris['target']== i]['pca_y']
  ax.scatter(x_axis_data,y_axis_data,marker=markers[i],label=iris_dataset.get('target_names')[i])

ax.legend()
plt.show()
```

##### 각 주성분이 분석에서 차지하는 비율

- .explained_variance_ratio_ : 주성분의 개수만큼 비율이 출력된다.

    ex) PCA(2) -> 2개의 값 출력

- 각 주성분이 차지하는 비율의 합이 95%를 넘어설 경우, 분석에 필요한 주성분 개수가 충분하다고 판단한다.

```python
print(iris_scale_pca.explained_variance_ratio_) 
```

- 점을 나타내는데 첫번째 주성분이 약 73% 관여했고, 두번째 주성분이 약23% 관여했다고 볼 수 있다.
- 각 주성분이 0.72962445, 0.22850762 일 경우, 분석에서 두 주성분이 차지하는 비율은 약 96% 이므로 충분하다.

## k-means

#### 군집분석
- 비지도학습의 한 가지
- 동질성 있는 데이터는 최대한 모으로, 이질성 있는 데이터는 최대한 멀리 계사하는 것이 군집분석의 목표
- 변수 값들의 유사성(거리)에 따라, 전체 데이터를 n개의 집단으로 분류하는 분석기법

#### k-means
- k개의 군집을 만드는 알고리즘
- 군집의 중심이 될 데이터 k개를 무작위(random)로 선별하고, 임의의 값 혹은 더 이상 중심으로 결정할 데이터가 없을 때까지 반복
- 데이터 분포가 군집에 적합하지 않을 경우, 제대로 동작되지 않음

#### 붓꽃 데이터의 군집화

- n_cluster: 생성할 군집의 개수(기본값=8)
- init: 군집화 알고리즘(기본값='k-means++')

```python
from sklearn.cluster import KMeans
k_mean=KMeans(n_clusters=3,init='k-means++') # 3개의 군집 생성, k-means++ 방법 사용
k_mean.fit(df_iris[feature_names])
```

```python
k_mean.labels_ # kmeans 알고리즘을 통해 군집화된 결과
```

```python
k_mean.labels_.shape # 데이터셋의 행의 개수와 같은 것을 확인할 수 있다.
```

#### 군집화 결과를 새로운 열로 정의

```python
df_iris['kmean_cluster']=k_mean.labels_ # kmean_cluster 열 추가
df_iris
```

#### 기존 분류 결과인 열 target과 k-means 알고리즘으로 분류한 결과인 열 kmean_cluster 비교

```python
df_iris.groupby(['target','kmean_cluster'])['sepal_length'].count()
df_iris.groupby(['target','kmean_cluster'])['sepal_width'].count()
df_iris.groupby(['target','kmean_cluster'])['petal_length'].count()
df_iris.groupby(['target','kmean_cluster'])['petal_width'].count()
```

군집화된 결과를 보면, 
- target에서의 0번과 kmean_cluster의 0번의 개수가 같으므로, 0번은 모두 잘 분류되었다.
- target에서 1번으로 분류된 것은 kmean_cluster에서는 1,2번으로 분류되었다.
  - 2개의 데이터가 2번으로 잘못 분류되었다.
- target에서 2번으로 분류된 것은 kmean_cluster에서 1,2번으로 분류되었다.
  - 14개의 데이터가 1번으로 잘못 분류되었다.

#### k-means의 군집 개수는 3개가 최적일까?
- 붓꽃 데이터에 이미 3개의 붓꽃 종류로 분류되어있음 -> 지도 학습에 최적
- 만약 붓꽃 종류가 분류되어 있지 않다면, 군집 개수가 3개가 적당한가? -> 비지도 학습의 목표

#### Elbow Method(엘보우 기법)
- 군집의 개수가 늘어날수록, 각 군집에서 중심-값 간 거리 제곱의 총 합은 줄어듦.
- 중심-값 같 거리 제곱의 총 합이 급격하게 줄어들다가 완만해진다면, 더 이상 군집을 나눌 필요가 없음
  - 부적합-과적합의 일종

##### Elbow Method는 k-means를 여러 번 진행하여 비교

- inertia_ : 군집의 중심-값 간 거리 제곱의 합을 제공
- 결과 그래프 모양이 팔꿈치와 비슷하다고 해서 Elbow Method
- 급격하게 완만해지는 지점의 군집의 개수가 적당하다고 볼 수 있다.
  - 아래 그래프에서는 3개 혹은 4개 지점

```python
elbow_inertias=list()

for i in range(1,10):
  k_mean_elbow=KMeans(n_clusters=i, init='random',n_init=15,max_iter=150,random_state=0)
  k_mean_elbow.fit(df_iris[feature_names])
  elbow_inertias.append(k_mean_elbow.inertia_) # intertia_ : 군집의 중심-값 간 거리 제곱의 합

plt.plot(range(1,10),elbow_inertias,marker='o')
plt.xlabel('Number of Cluster') # x축: 군집의 개수
plt.ylabel('Sum of squared distances') # y축: 중심-값 간 거리의 누적 제곱 합
plt.show()
```
